El dataframe contienen 30 columnas, que permiten saber fechas ( año, mes y dia ) sobre vuelos aereos. tambien nos brinda informacion exacta por vuelo, salida del vuelo, horario computarizado de salida de vuelo, hora de llegada, informacion de la compañia por cada vuelo, maniobras del avion en tierra, cancelaciones, codigos de calcelacion, desviacion del vuelo y ademas nos clasifica los motivos de retraso de vuelos ( Operadores, clima, NAS, Security, Retraso acumulado ).

Las columnas "DepTime", "CRSDepTime", "	ArrTime", "CRSArrTime", contienen una hora especifica.

El resto de columnas contienen datos en minutos. 

las columnas que clasifican los retrasos, contienen datos Nan, por que dependiendo el tipo de retraso se vera reflejado en la columna ArrDelay, una relacion encontrada es que cuando haya un registro en la columna ( LateAircraftDelay ) habra un registro en la columna (CarrierDelay), esto se suma y aparecera el total en ArrDelay.

CarrierDelay= Retraso por el operador del avion.

LateAircraftDelay= Retraso por que el avion ya venia de un vuelo anterior con retraso. 

taxiIn= Rodaje del avion desde la pista hasta el lugar de estacionamiento.

taxiout= Rodaje del avion hacia la pista, para iniciar vuelo.

SecurityDelaY = Retraso por cuestiones de seguridad.

NASDelay = Retrasos causados por el National Airspace System.

WeatherDelay = Retraso por clima

Tambien se identifica que:

AirTime + taxiIn + taxiOut= ActualElapsedTime (minutos), duracion del vuelo.

DepTime (dato de hora) - ArrTime (dato de hora) = ActualElapsedTime (minutos), duracion del vuelo.

DepTime (hora dada) + ActualElapsedTime (minutos)= ArrTime (hora de llegada real).

ArrTime (minutos,real) - ActualElapsedTime (minutos) = DepTime (hora dde salida real).

AirTime + ArrDelay + DepDelay + taxiIn + taxiOut= CRSElapsedTime (minutos) (computarizado).

DepTime - CRSDepTime = DepDelay (retraso que sufrio el vuelo a  su salida).

ArrTime - CRSArrTime = ArrDelay (retraso que sufrio el vuelo a  su llegada).

CRSArrTime - ArrDelay = ArrTime (Hora de llegada del vuelo).

```python
df3['ClassArrDelay'] = df3['ArrDelay'].apply(lambda x: 'Tarde' if x >= 1.0 else 'A tiempo')
df3
```

```python
df3['WhyDelay'] = df3[['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']].apply(
    lambda row: 'undefined' if all(val == 0 for val in row) else row.idxmax(),
    axis=1)
df3
```
